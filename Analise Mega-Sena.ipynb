```python
#Declaração das Bibliotecas a serem importadas

import matplotlib as plt
import pandas as pd

#Lendo os dados do arquivo e armazenando na variável leitor
data = 'basedados.xlsx'
excel = pd.ExcelFile(data)
print(excel.sheet_names)
leitor = excel.parse(0)
leitor.head()
```

```python
#Vendo a quantidade de concursos realizados
leitor.shape[0]
```

```python
#Criando uma lista com os nomes das colunas que armazenam os números sorteados 
colunas = ["Dezena1", "Dezena2", "Dezena3", "Dezena4", "Dezena5", "Dezena6"]

#Observando os 6 números que mais apareceram nos decorrer de todos os sorteios 
leitor[colunas].unstack().value_counts()[0:6]
```

```python
#Análise dos 50 primeiros Sorteios
leitor.iloc[0:50]
```

```python
#Vendo os números que mais apareceram no decorrer do 50 primeiros sorteios
leitor[leitor["Concurso"] < 50][colunas].unstack().value_counts()[0:6]
```

```python
#Análise dos 50 últimos Sorteios
leitor.iloc[-51:-1]
```

```python
#Vendo os números que mais apareceram no decorrer do 50 últimos sorteios
leitor[leitor["Concurso"] > 1974][colunas].unstack().value_counts()[0:6]
```

```python
#Análise dos 10 primeiros Sorteios
leitor.iloc[-10:-1]
```

```python
#Análise dos 10 primeiros Sorteios, para focar nos que tem dado mais "sorte" ultimamente
leitor[leitor["Data Sorteio"] > "2018-02-21"][colunas].unstack().value_counts()[0:13]
```

```python
#Análise dos 5 primeiros Sorteios
leitor.iloc[-5:-1]
```

```python
#Análise dos 5 primeiros Sorteios, para focar nos que tem dado mais "sorte" ultimamente
leitor[leitor["Concurso"] > 2020][colunas].unstack().value_counts()[0:13]
```

```python
#Criando uma coluna para Armazenar os valor total do prêmio
leitor["Prêmio"] = leitor.Ganhadores_Sena * leitor.Rateio_Sena
leitor.head()
```

```python
#Vendo qual sorteio teve o maior prêmio
leitor.iloc[leitor.Prêmio.idxmax()]
```

```python
#Criando uma coluna para saber o sorteio com o maior número de ganhadores
leitor["Total de Ganhadores"] = leitor["Ganhadores_Sena"] + leitor["Ganhadores_Quina"] + leitor["Ganhadores_Quadra"]
leitor.head()
```

```python
#Vendo qual sorteio teve o maior sorteio
leitor.iloc[leitor["Total de Ganhadores"].idxmax()]
```

```python
#As próximas três analisarão quais sorteio tem o maior número de ganhadores em cada possibilidade de vitória
leitor.iloc[leitor["Ganhadores_Quina"].idxmax()]
```
